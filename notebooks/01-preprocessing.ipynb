## Cleaning

### Missing values

```python
# Drop rows with missing values
df = df.dropna()
print(df.shape)
```

### Unimportant columns

```python
# Drop customer id column
df = df.drop('CUST_ID', axis=1)
```

### Outliers 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of columns for which we will plot histograms
cols = ['BALANCE', 'PURCHASES', 'ONEOFF_PURCHASES', 'CREDIT_LIMIT', 'PAYMENTS']

for col in cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True)
    plt.title('Histogram of ' + col)
    plt.show()
```

```python
# For each column, remove the rows with outliers
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:  # Only for numeric columns
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        filter = (df[col] >= Q1 - 1.5 * IQR) & (df[col] <= Q3 + 1.5 * IQR)
        df_no_outliers = df.loc[filter]

print(f"Remaining rows in the dataset: {len(df_no_outliers)}")
```

```python
# Save the processed dataset
df_no_outliers.to_csv('/workspaces/credit-mlops/data/processed/processed_data.csv', index=False)
```
